```python
import matplotlib.pyplot as plt
import numpy as np
from image_utils import load_image, split_image, display_patches_with_indices, detect_grid_size
```

```python
# Path to your image - REPLACE THIS with your actual image path
image_path = "./Gravity Falls/puzzle_4x4/6.jpg" 

# Detect grid size from image content (gradients)
grid_size = detect_grid_size(image_path)
print(f"Detected grid size: {grid_size}x{grid_size}")
```

```python
if grid_size is not None:
    # Load the image
    image = load_image(image_path)

    if image is not None:
        print(f"Image loaded. Shape: {image.shape}")
        
        # Split into patches
        patches = split_image(image, grid_size)
        print(f"Number of patches: {len(patches)}")
        
        # Display patches with indices
        display_patches_with_indices(patches, grid_size)
    else:
        print("Please check the image path.")
else:
    print("Could not detect grid size from image content.")
```
